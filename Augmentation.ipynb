```python
from audiomentations import Compose, AddGaussianNoise, TimeStretch, PitchShift, Shift,LoudnessNormalization,AddGaussianSNR
import numpy as np
from scipy.io import wavfile
import matplotlib.pyplot as plt
```

```python
RAW_DATASET_PATH="/home/zhengxio/Xingyu/Covid_data/IEEE_HealthCareSummit_Dev_Data_Release"
```

```python
def get_labels():
    label_covid  = {}
    label_gender = {} # not used 
    postives=[]
    negatives=[]
    ids=[]
    with open(RAW_DATASET_PATH+"/metadata.csv") as f:
        next(f) # remove header
        for line in f:
            splits = line.replace("\n","").split(" ")
            voice_id = splits[0]
            label_gender[voice_id]=splits[2] # not used
            label_covid[voice_id]=0 if splits[1] == "n" else 1
            ids.append(voice_id)
            if splits[1] == "n":
                negatives.append(voice_id)
            else:
                postives.append(voice_id)
    return label_covid,ids,postives,negatives
```

```python

augment = Compose([
    # LoudnessNormalization(p=1.0),
    AddGaussianSNR(min_snr_in_db=0, max_snr_in_db=35, p=1.0),
    AddGaussianNoise(min_amplitude=0.001, max_amplitude=0.015, p=0.5),
    # TimeStretch(min_rate=0.1, max_rate=1.25, p=0.5),
    PitchShift(min_semitones=-4, max_semitones=4, p=0.5),
    Shift(min_fraction=-0.5, max_fraction=0.5, p=0.5),
])

def augemnt_audio(data,fs):
    return augment(samples=data, sample_rate=fs)
```

```python
label_covid,ids,p_lists,n_lists= get_labels()

```

```python
for t in ["breathing","cough","speech"]:
    AUDIO_PATH=RAW_DATASET_PATH+"/AUDIO_WAV/"+t 
    for ID in ids:
        n_aug = 1 if label_covid[ID] == 0 else 7
        fs, data = wavfile.read(AUDIO_PATH+"/"+ID+".wav")
        for aug_id in range(n_aug):
            print("{} {} {}".format(t,ID,aug_id))
            aug_data = augemnt_audio(data.astype(np.float64),fs)
            wavfile.write(AUDIO_PATH+"/"+ID+"_{}.wav".format(aug_id+1), fs, aug_data.astype(np.int16))
        
```
